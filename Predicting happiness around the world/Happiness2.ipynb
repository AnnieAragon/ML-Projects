## Statsmodels:

```python
import statsmodels.api as sm
```

### Feature selection: Happiness report 6 variables

```python
features = [
'health',
'social_support',
'gdp_h',
'freedom',
'generosity',
'government_trust',
# 'phones',
# 'literacy',
# 'Infant_mortality'
]

x= features
y = 'happiness_score'

train, test = train_test_split(df1, train_size=0.7, random_state= 1)

df1train = pd.DataFrame(train, columns=df1.columns)
df1test = pd.DataFrame(test, columns=df1.columns)


x = df1train[x]
x = sm.add_constant(x)
y = df1train[y]

model = sm.OLS(y, x).fit()
print(model.summary())

xt = df1test[features]
xt = sm.add_constant(xt)
yt = df1test['happiness_score']

print(r2_score(yt, model.predict(xt)))
```
